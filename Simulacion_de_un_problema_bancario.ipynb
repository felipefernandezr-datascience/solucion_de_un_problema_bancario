```python
import random
import heapq
from dataclasses import dataclass, field
import statistics

# ---------------------------
# CONFIGURACIÓN DEL MODELO
# ---------------------------

DURACION_DIA = 480  # minutos
NUM_CAJEROS = 3
N_REPLICAS = 10


# Probabilidades de usuario
probs_tipo_accion = {
    "retiro": 0.70,
    "pago": 0.30
}

probs_usuario = {
    "retiro": {
        "rapido": 0.23,
        "normal": 0.40,
        "lento": 0.17,
        "muy_lento": 0.20
    },
    "pago": {
        "rapido": 0.10,
        "normal": 0.20,
        "lento": 0.30,
        "muy_lento": 0.40
    }
}


# Tiempos de llegada
llegada_media = {
    "retiro": {
        "rapido": 1,
        "normal": 2,
        "lento": 3,
        "muy_lento": 3
    },
    "pago": {
        "rapido": 1,
        "normal": 2,
        "lento": 3,
        "muy_lento": 4
    }
}


# Tiempos de servicio
servicio_media = {
    "retiro": {
        "rapido": 1,
        "normal": 2,
        "lento": 3,
        "muy_lento": 4
    },
    "pago": {
        "rapido": 3,
        "normal": 3,
        "lento": 5,
        "muy_lento": 7
    }
}

# ---------------------------
# ESTRUCTURAS DE DATOS
# ---------------------------

@dataclass
class Cajero:
    ocupado: bool = False
    cola: list = field(default_factory=list)
    tiempo_total_servicio: float = 0
    total_atendidos: int = 0
    tiempos_espera: list = field(default_factory=list)


# Selecciones probabilísticas ----------------
def seleccionar_accion():
    return random.choices(["retiro", "pago"], weights=[0.70, 0.30])[0]

def seleccionar_tipo_usuario(accion):
    tipos = list(probs_usuario[accion].keys())
    pesos = list(probs_usuario[accion].values())
    return random.choices(tipos, weights=pesos)[0]

# Generación de tiempos exponenciales ----------
def generar_tiempo_exponencial(media):
    return random.expovariate(1/media)


# ---------------------------
# SIMULACIÓN DE UNA RÉPLICA
# ---------------------------

def simular_un_dia():
    reloj = 0
    cajeros = [Cajero() for _ in range(NUM_CAJEROS)]
    eventos = []  # (tiempo, tipo_evento, cajero_id, datos_usuario)

    # Programar primera llegada
    accion = seleccionar_accion()
    tipo = seleccionar_tipo_usuario(accion)
    tiempo_llegada = generar_tiempo_exponencial(llegada_media[accion][tipo])

    heapq.heappush(eventos, (tiempo_llegada, "llegada", None, (accion, tipo, tiempo_llegada)))

    # Datos por tipo
    conteo_tipos = {"retiro": 0, "pago": 0}

    while eventos:
        tiempo, evento, id_cajero, datos = heapq.heappop(eventos)
        reloj = tiempo

        if reloj > DURACION_DIA:
            break

        if evento == "llegada":
            accion, tipo, momento = datos
            conteo_tipos[accion] += 1

            # Asignar al cajero con menor cola
            id_cajero = min(range(NUM_CAJEROS), key=lambda i: len(cajeros[i].cola))
            cajero = cajeros[id_cajero]

            llegada = reloj

            if not cajero.ocupado:
                cajero.ocupado = True
                tiempo_serv = generar_tiempo_exponencial(servicio_media[accion][tipo])
                fin_servicio = reloj + tiempo_serv
                heapq.heappush(eventos, (fin_servicio, "fin_servicio", id_cajero, llegada))
                cajero.tiempo_total_servicio += tiempo_serv
            else:
                cajero.cola.append(llegada)

            # Programar próxima llegada
            accion2 = seleccionar_accion()
            tipo2 = seleccionar_tipo_usuario(accion2)
            tiempo_llegada = reloj + generar_tiempo_exponencial(llegada_media[accion2][tipo2])
            heapq.heappush(eventos, (tiempo_llegada, "llegada", None, (accion2, tipo2, tiempo_llegada)))

        elif evento == "fin_servicio":
            cajero = cajeros[id_cajero]
            llegada_usuario = datos
            cajero.total_atendidos += 1
            cajero.ocupado = False

            if cajero.cola:
                llegada_next = cajero.cola.pop(0)
                cajero.tiempos_espera.append(reloj - llegada_next)
                tiempo_serv = generar_tiempo_exponencial(2)  # corregir según acción/tipo si deseas más detalle
                fin_servicio = reloj + tiempo_serv
                heapq.heappush(eventos, (fin_servicio, "fin_servicio", id_cajero, llegada_next))
                cajero.ocupado = True

    return cajeros, conteo_tipos


# ---------------------------
# EJECUCIÓN DE LAS RÉPLICAS
# ---------------------------

def ejecutar_simulacion():
    resultados = []

    for r in range(N_REPLICAS):
        cajeros, conteo = simular_un_dia()
        resultados.append((cajeros, conteo))

    return resultados


# -------- Ejecutar ---------
if __name__ == "__main__":
    resultados = ejecutar_simulacion()
    print("Simulación completada.")
```

```python
import matplotlib.pyplot as plt

# ============================================
#  ANÁLISIS + GRÁFICAS PARA EL INFORME
# ============================================

def generar_graficas(resultados):
    """
    Genera todas las gráficas explicativas para mostrar los resultados
    solicitados en el trabajo.
    """

    # ----------------------------
    # 1. Tiempo promedio de atención por cajero
    # ----------------------------
    tiempos_promedio_cajeros = []

    for cajero_id in range(NUM_CAJEROS):
        tiempos = []
        for replica in resultados:
            cajeros, _ = replica
            c = cajeros[cajero_id]
            if c.total_atendidos > 0:
                tiempos.extend([c.tiempo_total_servicio / c.total_atendidos])
        tiempos_promedio_cajeros.append(statistics.mean(tiempos))

    plt.figure(figsize=(8, 5))
    plt.bar(range(1, NUM_CAJEROS + 1), tiempos_promedio_cajeros)
    plt.title("Tiempo promedio de atención por cajero")
    plt.xlabel("Cajero")
    plt.ylabel("Tiempo promedio (min)")
    plt.xticks(range(1, NUM_CAJEROS + 1))
    plt.grid(axis="y", linestyle="--", alpha=0.5)
    plt.show()

    # ----------------------------
    # 2. Promedio de usuarios por tipo (retiro / pago)
    # ----------------------------
    totales_retiro = []
    totales_pago = []

    for cajeros, conteo in resultados:
        totales_retiro.append(conteo["retiro"])
        totales_pago.append(conteo["pago"])

    prom_retiros = statistics.mean(totales_retiro)
    prom_pagos = statistics.mean(totales_pago)

    plt.figure(figsize=(6, 6))
    plt.pie([prom_retiros, prom_pagos],
            labels=["Retiros", "Pagos"],
            autopct="%1.1f%%",
            startangle=90)
    plt.title("Porcentaje promedio de usuarios por tipo")
    plt.show()

    # ----------------------------
    # 3. Usuarios por tipo en cada réplica
    # ----------------------------
    plt.figure(figsize=(10, 5))
    plt.plot(totales_retiro, marker="o", label="Retiros")
    plt.plot(totales_pago, marker="s", label="Pagos")
    plt.title("Usuarios por tipo en cada réplica")
    plt.xlabel("Número de réplica")
    plt.ylabel("Número de usuarios")
    plt.grid(True, linestyle="--", alpha=0.5)
    plt.legend()
    plt.show()

    # ----------------------------
    # 4. Tiempos promedio de espera → ¿nuevo cajero?
    # ----------------------------
    tiempos_espera_todos = []
    for cajeros, _ in resultados:
        for c in cajeros:
            tiempos_espera_todos.extend(c.tiempos_espera)

    if tiempos_espera_todos:
        prom_espera = statistics.mean(tiempos_espera_todos)
    else:
        prom_espera = 0

    plt.figure(figsize=(7, 5))
    plt.hist(tiempos_espera_todos, bins=12)
    plt.title(f"Distribución de tiempos de espera\nPromedio = {prom_espera:.2f} min")
    plt.xlabel("Tiempo de espera (min)")
    plt.ylabel("Frecuencia")
    plt.grid(True, linestyle="--", alpha=0.5)
    plt.show()

    # ----------------------------
    # 5. Decidir cuántos cajeros deben ser exclusivos
    # ----------------------------
    # Criterio: comparar carga de trabajo por tipo
    porc_retiros = [r / (r + p) for r, p in zip(totales_retiro, totales_pago)]
    porc_pagos = [p / (r + p) for r, p in zip(totales_retiro, totales_pago)]

    plt.figure(figsize=(10, 5))
    plt.plot(porc_retiros, marker="o", label="Porcentaje retiros", color="blue")
    plt.plot(porc_pagos, marker="s", label="Porcentaje pagos", color="green")
    plt.title("Proporción de carga por tipo de usuario en cada réplica")
    plt.xlabel("Réplica")
    plt.ylabel("Proporción")
    plt.grid(True, linestyle="--", alpha=0.5)
    plt.legend()
    plt.show()

    print("\n=== RECOMENDACIÓN DE CAJEROS EXCLUSIVOS ===")
    prom_retiros_prop = statistics.mean(porc_retiros)
    prom_pagos_prop = statistics.mean(porc_pagos)
    print(f"Promedio retiros: {prom_retiros_prop:.2f}")
    print(f"Promedio pagos:   {prom_pagos_prop:.2f}")

    if prom_retiros_prop > 0.65:
        print("→ Se recomienda al menos 1 cajero exclusivo para RETIROS.")
    if prom_pagos_prop > 0.40:
        print("→ Se recomienda 1 cajero exclusivo para PAGOS.")
    print("Decisión final depende del nivel de servicio requerido.")


# ============================================
# EJECUCIÓN COMPLETA (Simulación + Gráficas)
# ============================================

if __name__ == "__main__":
    resultados = ejecutar_simulacion()
    procesar_resultados(resultados)
    generar_graficas(resultados)
```

```python
import random
import heapq
import statistics
import matplotlib.pyplot as plt
from dataclasses import dataclass, field

# ---------------------------
# 1. CONFIGURACIÓN DEL MODELO
# ---------------------------

DURACION_DIA = 480  # minutos
NUM_CAJEROS = 3
N_REPLICAS = 10

# Probabilidades de usuario
probs_tipo_accion = {
    "retiro": 0.70,
    "pago": 0.30
}

probs_usuario = {
    "retiro": { "rapido": 0.23, "normal": 0.40, "lento": 0.17, "muy_lento": 0.20 },
    "pago":   { "rapido": 0.10, "normal": 0.20, "lento": 0.30, "muy_lento": 0.40 }
}

# Tiempos de llegada (media exponencial)
llegada_media = {
    "retiro": { "rapido": 1, "normal": 2, "lento": 3, "muy_lento": 3 },
    "pago":   { "rapido": 1, "normal": 2, "lento": 3, "muy_lento": 4 }
}

# Tiempos de servicio (media exponencial)
servicio_media = {
    "retiro": { "rapido": 1, "normal": 2, "lento": 3, "muy_lento": 4 },
    "pago":   { "rapido": 3, "normal": 3, "lento": 5, "muy_lento": 7 }
}

# ---------------------------
# 2. ESTRUCTURAS DE DATOS
# ---------------------------

@dataclass
class Cajero:
    tipo_cajero: str # "Retiros" o "Pagos"
    ocupado: bool = False
    cola: list = field(default_factory=list)
    tiempo_total_servicio: float = 0
    total_atendidos: int = 0
    tiempos_espera: list = field(default_factory=list)

def seleccionar_accion():
    return random.choices(["retiro", "pago"], weights=[0.70, 0.30])[0]

def seleccionar_tipo_usuario(accion):
    tipos = list(probs_usuario[accion].keys())
    pesos = list(probs_usuario[accion].values())
    return random.choices(tipos, weights=pesos)[0]

def generar_tiempo_exponencial(media):
    return random.expovariate(1/media)

# ---------------------------
# 3. SIMULACIÓN (LÓGICA MODIFICADA)
# ---------------------------

def simular_un_dia_exclusivo():
    reloj = 0
    
    # DEFINICIÓN DE ROLES DE CAJEROS
    # Cajero 0: Exclusivo Retiros
    # Cajero 1 y 2: Exclusivos Pagos
    cajeros = [
        Cajero(tipo_cajero="Exclusivo Retiro"), # ID 0
        Cajero(tipo_cajero="Exclusivo Pago"),   # ID 1
        Cajero(tipo_cajero="Exclusivo Pago")    # ID 2
    ]
    
    eventos = []  # (tiempo, tipo_evento, cajero_id, datos_usuario)

    # Programar primera llegada
    accion = seleccionar_accion()
    tipo = seleccionar_tipo_usuario(accion)
    tiempo_llegada = generar_tiempo_exponencial(llegada_media[accion][tipo])
    
    heapq.heappush(eventos, (tiempo_llegada, "llegada", None, (accion, tipo, tiempo_llegada)))

    conteo_tipos = {"retiro": 0, "pago": 0}

    while eventos:
        tiempo, evento, id_cajero, datos = heapq.heappop(eventos)
        reloj = tiempo

        if reloj > DURACION_DIA:
            break

        if evento == "llegada":
            accion, tipo, momento = datos
            conteo_tipos[accion] += 1

            # --- LÓGICA DE ASIGNACIÓN EXCLUSIVA ---
            id_seleccionado = -1
            
            if accion == "retiro":
                # Solo puede ir al Cajero 0
                id_seleccionado = 0
                
            else: # accion == "pago"
                # Puede ir al Cajero 1 o 2 (el que tenga menos cola)
                # Comparamos len(cola) de cajeros[1] y cajeros[2]
                opciones_pago = [1, 2]
                id_seleccionado = min(opciones_pago, key=lambda i: len(cajeros[i].cola))

            cajero = cajeros[id_seleccionado]
            llegada = reloj

            if not cajero.ocupado:
                cajero.ocupado = True
                tiempo_serv = generar_tiempo_exponencial(servicio_media[accion][tipo])
                fin_servicio = reloj + tiempo_serv
                heapq.heappush(eventos, (fin_servicio, "fin_servicio", id_seleccionado, llegada))
                cajero.tiempo_total_servicio += tiempo_serv
            else:
                cajero.cola.append(llegada)

            # Programar próxima llegada
            accion2 = seleccionar_accion()
            tipo2 = seleccionar_tipo_usuario(accion2)
            tiempo_llegada = reloj + generar_tiempo_exponencial(llegada_media[accion2][tipo2])
            heapq.heappush(eventos, (tiempo_llegada, "llegada", None, (accion2, tipo2, tiempo_llegada)))

        elif evento == "fin_servicio":
            cajero = cajeros[id_cajero]
            llegada_usuario = datos
            cajero.total_atendidos += 1
            cajero.ocupado = False

            if cajero.cola:
                llegada_next = cajero.cola.pop(0)
                cajero.tiempos_espera.append(reloj - llegada_next)
                
                # Para calcular el tiempo de servicio del siguiente, necesitamos saber qué es.
                # Como son colas exclusivas, deducimos la acción por el tipo de cajero
                if cajero.tipo_cajero == "Exclusivo Retiro":
                    accion_next = "retiro"
                else:
                    accion_next = "pago"
                
                # Asumimos un tipo 'normal' para simplificar o deberíamos haber guardado el tipo en la cola
                # Para mantener consistencia con tu código original que usaba media=2 genérica en fin_servicio:
                # Usaremos la media ponderada aproximada o generaremos un tipo aleatorio nuevo:
                tipo_next = seleccionar_tipo_usuario(accion_next) 
                tiempo_serv = generar_tiempo_exponencial(servicio_media[accion_next][tipo_next])
                
                fin_servicio = reloj + tiempo_serv
                heapq.heappush(eventos, (fin_servicio, "fin_servicio", id_cajero, llegada_next))
                cajero.ocupado = True
                cajero.tiempo_total_servicio += tiempo_serv

    return cajeros, conteo_tipos

def ejecutar_simulacion():
    resultados = []
    for r in range(N_REPLICAS):
        cajeros, conteo = simular_un_dia_exclusivo()
        resultados.append((cajeros, conteo))
    return resultados

# ---------------------------
# 4. ANÁLISIS Y GRÁFICAS
# ---------------------------

def generar_graficas(resultados):
    # 1. Tiempo promedio de atención por cajero
    tiempos_promedio_cajeros = []
    etiquetas_cajeros = ["C1 (Retiro)", "C2 (Pago)", "C3 (Pago)"] # Etiquetas personalizadas

    for cajero_id in range(NUM_CAJEROS):
        tiempos = []
        for replica in resultados:
            cajeros, _ = replica
            c = cajeros[cajero_id]
            if c.total_atendidos > 0:
                tiempos.append(c.tiempo_total_servicio / c.total_atendidos)
            else:
                tiempos.append(0)
        tiempos_promedio_cajeros.append(statistics.mean(tiempos) if tiempos else 0)

    plt.figure(figsize=(8, 5))
    barras = plt.bar(range(NUM_CAJEROS), tiempos_promedio_cajeros, color=['blue', 'orange', 'orange'])
    plt.title("Tiempo promedio de atención por cajero (Config: 1 Retiro, 2 Pagos)")
    plt.xlabel("Cajero y Asignación")
    plt.ylabel("Tiempo promedio (min)")
    plt.xticks(range(NUM_CAJEROS), etiquetas_cajeros)
    plt.grid(axis="y", linestyle="--", alpha=0.5)
    plt.show()

    # 2. Tiempos promedio de espera (General)
    tiempos_espera_todos = []
    for cajeros, _ in resultados:
        for c in cajeros:
            tiempos_espera_todos.extend(c.tiempos_espera)

    prom_espera = statistics.mean(tiempos_espera_todos) if tiempos_espera_todos else 0

    plt.figure(figsize=(7, 5))
    plt.hist(tiempos_espera_todos, bins=15, color='purple', alpha=0.7)
    plt.title(f"Distribución General Tiempos de Espera\nPromedio Global = {prom_espera:.2f} min")
    plt.xlabel("Tiempo de espera (min)")
    plt.ylabel("Frecuencia")
    plt.grid(True, linestyle="--", alpha=0.5)
    plt.show()

    # 3. Comparación Espera: Retiros vs Pagos
    esperas_retiro = []
    esperas_pago = []
    
    for cajeros, _ in resultados:
        # Cajero 0 es retiro
        esperas_retiro.extend(cajeros[0].tiempos_espera)
        # Cajeros 1 y 2 son pago
        esperas_pago.extend(cajeros[1].tiempos_espera)
        esperas_pago.extend(cajeros[2].tiempos_espera)

    prom_esp_retiro = statistics.mean(esperas_retiro) if esperas_retiro else 0
    prom_esp_pago = statistics.mean(esperas_pago) if esperas_pago else 0

    print("\n=== RESULTADOS DE CONFIGURACIÓN (1 RETIRO / 2 PAGOS) ===")
    print(f"Tiempo medio espera cola RETIROS (C1): {prom_esp_retiro:.2f} min")
    print(f"Tiempo medio espera cola PAGOS (C2, C3): {prom_esp_pago:.2f} min")
    
    # Decisión rápida basada en la corrida
    if abs(prom_esp_retiro - prom_esp_pago) > 2:
        print("\nOBSERVACIÓN: Hay un desbalance significativo entre las colas.")
        if prom_esp_retiro > prom_esp_pago:
            print("-> La cola de RETIROS está sufriendo.")
        else:
            print("-> La cola de PAGOS está sufriendo.")
    else:
        print("\nOBSERVACIÓN: El balance parece aceptable.")

# ---------------------------
# EJECUCIÓN
# ---------------------------

if __name__ == "__main__":
    print("Iniciando simulación con configuración: 1 Cajero Retiros / 2 Cajeros Pagos...")
    resultados = ejecutar_simulacion()
    generar_graficas(resultados)
```

```python
import random
import heapq
import statistics
import matplotlib.pyplot as plt
from dataclasses import dataclass, field

# ==========================================
# CONFIGURACIÓN (Actualizada a 4 Cajeros)
# ==========================================
DURACION_DIA = 480 
NUM_CAJEROS = 4  # <--- CAMBIO PRINCIPAL
N_REPLICAS = 10

probs_tipo_accion = {"retiro": 0.70, "pago": 0.30}

probs_usuario = {
    "retiro": { "rapido": 0.23, "normal": 0.40, "lento": 0.17, "muy_lento": 0.20 },
    "pago":   { "rapido": 0.10, "normal": 0.20, "lento": 0.30, "muy_lento": 0.40 }
}

llegada_media = {
    "retiro": { "rapido": 1, "normal": 2, "lento": 3, "muy_lento": 3 },
    "pago":   { "rapido": 1, "normal": 2, "lento": 3, "muy_lento": 4 }
}

servicio_media = {
    "retiro": { "rapido": 1, "normal": 2, "lento": 3, "muy_lento": 4 },
    "pago":   { "rapido": 3, "normal": 3, "lento": 5, "muy_lento": 7 }
}

# ==========================================
# ESTRUCTURAS
# ==========================================
@dataclass
class Cajero:
    tipo_cajero: str 
    ocupado: bool = False
    cola: list = field(default_factory=list)
    tiempo_total_servicio: float = 0
    total_atendidos: int = 0
    tiempos_espera: list = field(default_factory=list)

def seleccionar_accion():
    return random.choices(["retiro", "pago"], weights=[0.70, 0.30])[0]

def seleccionar_tipo_usuario(accion):
    tipos = list(probs_usuario[accion].keys())
    pesos = list(probs_usuario[accion].values())
    return random.choices(tipos, weights=pesos)[0]

def generar_tiempo_exponencial(media):
    return random.expovariate(1/media)

# ==========================================
# LÓGICA DE SIMULACIÓN (2 RETIROS / 2 PAGOS)
# ==========================================

def simular_dia_4cajeros():
    reloj = 0
    
    # CONFIGURACIÓN DE 4 CAJEROS
    # Cajeros 0 y 1: Retiros
    # Cajeros 2 y 3: Pagos
    cajeros = [
        Cajero(tipo_cajero="Exclusivo Retiro"), # ID 0
        Cajero(tipo_cajero="Exclusivo Retiro"), # ID 1
        Cajero(tipo_cajero="Exclusivo Pago"),   # ID 2
        Cajero(tipo_cajero="Exclusivo Pago")    # ID 3
    ]
    
    eventos = [] 

    # Primera llegada
    accion = seleccionar_accion()
    tipo = seleccionar_tipo_usuario(accion)
    tiempo_llegada = generar_tiempo_exponencial(llegada_media[accion][tipo])
    heapq.heappush(eventos, (tiempo_llegada, "llegada", None, (accion, tipo, tiempo_llegada)))

    conteo_tipos = {"retiro": 0, "pago": 0}

    while eventos:
        tiempo, evento, id_cajero, datos = heapq.heappop(eventos)
        reloj = tiempo

        if reloj > DURACION_DIA:
            break

        if evento == "llegada":
            accion, tipo, momento = datos
            conteo_tipos[accion] += 1

            # --- LÓGICA DE ENRUTAMIENTO (2 vs 2) ---
            id_seleccionado = -1
            
            if accion == "retiro":
                # Buscar min cola entre cajeros 0 y 1
                opciones = [0, 1]
                id_seleccionado = min(opciones, key=lambda i: len(cajeros[i].cola))
                
            else: # accion == "pago"
                # Buscar min cola entre cajeros 2 y 3
                opciones = [2, 3]
                id_seleccionado = min(opciones, key=lambda i: len(cajeros[i].cola))

            cajero = cajeros[id_seleccionado]
            llegada = reloj

            if not cajero.ocupado:
                cajero.ocupado = True
                tiempo_serv = generar_tiempo_exponencial(servicio_media[accion][tipo])
                fin_servicio = reloj + tiempo_serv
                heapq.heappush(eventos, (fin_servicio, "fin_servicio", id_seleccionado, llegada))
                cajero.tiempo_total_servicio += tiempo_serv
            else:
                cajero.cola.append(llegada)

            # Próxima llegada
            accion2 = seleccionar_accion()
            tipo2 = seleccionar_tipo_usuario(accion2)
            tiempo_llegada = reloj + generar_tiempo_exponencial(llegada_media[accion2][tipo2])
            heapq.heappush(eventos, (tiempo_llegada, "llegada", None, (accion2, tipo2, tiempo_llegada)))

        elif evento == "fin_servicio":
            cajero = cajeros[id_cajero]
            llegada_usuario = datos
            cajero.total_atendidos += 1
            cajero.ocupado = False

            if cajero.cola:
                llegada_next = cajero.cola.pop(0)
                cajero.tiempos_espera.append(reloj - llegada_next)
                
                # Determinar tipo de servicio según el cajero
                if cajero.tipo_cajero == "Exclusivo Retiro":
                    accion_next = "retiro"
                else:
                    accion_next = "pago"
                
                tipo_next = seleccionar_tipo_usuario(accion_next)
                tiempo_serv = generar_tiempo_exponencial(servicio_media[accion_next][tipo_next])
                
                fin_servicio = reloj + tiempo_serv
                heapq.heappush(eventos, (fin_servicio, "fin_servicio", id_cajero, llegada_next))
                cajero.ocupado = True
                cajero.tiempo_total_servicio += tiempo_serv

    return cajeros, conteo_tipos

def ejecutar_simulacion():
    resultados = []
    for r in range(N_REPLICAS):
        cajeros, conteo = simular_dia_4cajeros()
        resultados.append((cajeros, conteo))
    return resultados

# ==========================================
# EVALUACIÓN DE NECESIDAD (GRÁFICAS)
# ==========================================

def generar_graficas_evaluacion(resultados):
    # 1. Tiempo promedio de atención
    tiempos_promedio_cajeros = []
    ocupacion_total = [] # Para medir % ocupación
    etiquetas_cajeros = ["C1(Ret)", "C2(Ret)", "C3(Pag)", "C4(Pag)"] 

    for cajero_id in range(NUM_CAJEROS):
        tiempos = []
        ocupaciones = []
        for replica in resultados:
            cajeros, _ = replica
            c = cajeros[cajero_id]
            
            # % de ocupación del día
            ocupaciones.append((c.tiempo_total_servicio / DURACION_DIA) * 100)
            
            if c.total_atendidos > 0:
                tiempos.append(c.tiempo_total_servicio / c.total_atendidos)
            else:
                tiempos.append(0)
        
        tiempos_promedio_cajeros.append(statistics.mean(tiempos) if tiempos else 0)
        ocupacion_total.append(statistics.mean(ocupaciones))

    # Gráfica de Ocupación (Vital para evaluar necesidad)
    plt.figure(figsize=(8, 5))
    barras = plt.bar(range(NUM_CAJEROS), ocupacion_total, color=['blue', 'blue', 'orange', 'orange'])
    plt.axhline(y=85, color='r', linestyle='--', label='Saturación (>85%)')
    plt.axhline(y=40, color='g', linestyle='--', label='Subutilización (<40%)')
    plt.title("Porcentaje de Ocupación por Cajero (4 Cajeros)")
    plt.xlabel("Cajero")
    plt.ylabel("% Tiempo Ocupado")
    plt.xticks(range(NUM_CAJEROS), etiquetas_cajeros)
    plt.legend()
    plt.grid(axis="y", linestyle="--", alpha=0.5)
    plt.show()

    # 2. Análisis comparativo de colas
    esperas_retiro = []
    esperas_pago = []
    
    for cajeros, _ in resultados:
        # Cajeros 0 y 1 son retiro
        esperas_retiro.extend(cajeros[0].tiempos_espera)
        esperas_retiro.extend(cajeros[1].tiempos_espera)
        # Cajeros 2 y 3 son pago
        esperas_pago.extend(cajeros[2].tiempos_espera)
        esperas_pago.extend(cajeros[3].tiempos_espera)

    prom_esp_retiro = statistics.mean(esperas_retiro) if esperas_retiro else 0
    prom_esp_pago = statistics.mean(esperas_pago) if esperas_pago else 0

    print("\n=== EVALUACIÓN DE NECESIDAD (4 CAJEROS: 2 RET / 2 PAG) ===")
    print(f"Tiempo medio espera RETIROS: {prom_esp_retiro:.2f} min")
    print(f"Tiempo medio espera PAGOS:   {prom_esp_pago:.2f} min")
    
    avg_ocupacion = statistics.mean(ocupacion_total)
    print(f"Ocupación promedio de la sucursal: {avg_ocupacion:.1f}%")

    # CONCLUSIÓN AUTOMÁTICA
    print("\n--- DICTAMEN ---")
    if prom_esp_retiro < 1 and prom_esp_pago < 1:
        print("RESULTADO: SOBREDIMENSIONADO.")
        print("Los tiempos de espera son insignificantes (< 1 min).")
        print("La 4ta caja NO es estrictamente necesaria y genera costos extra innecesarios.")
    elif avg_ocupacion < 50:
         print("RESULTADO: SUBUTILIZADO.")
         print("Los cajeros pasan más de la mitad del tiempo sin hacer nada.")
         print("Se recomienda mantenerse en 3 cajeros y optimizar.")
    else:
        print("RESULTADO: EFICIENTE.")
        print("Se han reducido los tiempos de espera a niveles aceptables sin desperdiciar demasiados recursos.")

if __name__ == "__main__":
    print("Evaluando escenario con 4 cajas...")
    resultados = ejecutar_simulacion()
    generar_graficas_evaluacion(resultados)
```
